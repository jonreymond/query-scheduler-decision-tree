```python
def create_plot_paths(path_time, ax, probas=None,):
    path_time_2 = sorted([(str(list(q_s)), t) for q_s, t in path_time])
    df_path = pd.DataFrame(path_time_2, columns=['variable','value'])
    df_path['process'] = "paths"
    axi = sns.histplot(data=df_path, hue = 'value', y='variable', weights='value',discrete=True, legend=False, ax=ax)
    for i, c in enumerate(ax.containers):

        labels = [v.get_width() if v.get_width() > 0 else '' for v in c]
        if probas is not None:
            probas = [p/sum(probas) for p in probas]
            labels = [str(round(labels[j], 2)) + ", Pr=" +str(round(probas[j], 3)) if str(labels[j]) != '' else '' for j in range(len(probas))] 

        axi.bar_label(c, labels=labels, label_type='center')
    
    if probas is None:
        ax.set_title("Sequential execution")
        ax.set(ylabel="paths")
    else:
        ax.set_title("Probabilistic execution")
        ax.set(ylabel="paths")
    # ax.xlabel('Time[s]');
```

```python
def get_col_queries(queries, res):
    data = {}
    t_total = 0
    for i,q in enumerate(queries) :
        s =  str(i + 1) + ".\n" + q + ":16\n"
        t = float(res[q](16))
        t_total += t
        data.update({s:[t]})
    return data, str(round(t_total, 3))

def batch_to_string(batch, index):
    t = batch[0]
    schedule = batch[1]
    s = str(index + 1) +".\n"
    for q,cores in schedule :
        s +=  str(q) + ":" + str(cores) + "\n"
    return s + "t:" + str(t), t

def get_col_schedule(res_schedule):
    data = {}

    for i, batch in enumerate(res_schedule):
        s, t = batch_to_string(batch, i)
        data.update({s:[t]})
    l_data = len(data)
        
    return data 
```

```python
def get_df(data, name_process):
    data.update({"processes": [name_process]})
    df = pd.DataFrame(data=data)

    df = df.melt(id_vars="processes")
    df.index = range(len(df))[::-1]
    df = df.iloc[::-1]
    return df
```
